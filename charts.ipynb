Подключим необходимые пакеты

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Загрузим и сгруппируем данные для mergeSort

```python
df = pd.read_csv('sort_results.csv')

merge_sort_data = df[df['sort_name'] == 'Merge Sort']

merge_sort_array_types = merge_sort_data['array_type'].unique()
```

Визуализируем данные

```python
plt.figure(figsize=(10, 6))

for array_type in merge_sort_array_types:
    array_data = merge_sort_data[merge_sort_data['array_type'] == array_type]
    plt.plot(array_data['size'], array_data['average_time'], label=f'{array_type} array')

plt.title('Merge Sort')
plt.xlabel('Array Size')
plt.ylabel('Average Time (ms)')
plt.legend()
plt.grid(True)
plt.show()
```

Хорошо заметно свойство сортировки слияниями: она работает одинаково во всех случаях.
